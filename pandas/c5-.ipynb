# nrows argument print some rows from file contains multiple rows

```python
# suppose i have a data in milloins billoins lines and i want to read some of them.
# then i have to use thie attribute  nrows=int

df1=pd.read_csv("D:\\8th_semester\\research data\\archive\\train.csv" , nrows=6)
print(df1)
```

# usecols=[ index of columns]  argument print some columns from file contains multiple columns

```python
# we have to pass the list as a value of usecols argument 
        # for printing first column usecols=[0]
        # for printing first and column usecols=[0,1]
        # for printing first second and 3rd column usecols=[0,1,2]

df2=pd.read_csv("D:\\8th_semester\\research data\\archive\\train.csv" , usecols=[0,1,3,5])
print(df2)
```

# skiprows=int parameter
# to make a next row as a heading of columns

```python
# stiprows=int we have to enter number how much rows we a=want to skip
# this function will strat rows skipping from first
df3=pd.read_csv("D:\\8th_semester\\research data\\archive\\train.csv" ,usecols=[0,1,2,3], skiprows=3)
print(df3)
```

# to skiprows=[index] any one or more row from file from anywhere

```python
# with the help of this parameter we can slip one or more rows form file from any where
# by passing a list as a value of skiprows
# index starts from zero
df4=pd.read_csv("D:\\8th_semester\\research data\\archive\\train.csv" ,usecols=[0,1,2,3], skiprows=[15,14,16,17,18])
# if we mastakenly add any not existing index this will not effect execution
print(df4)
```

# index_col="heading of column/or index number" creating any column as an index 

```python
# with column heading name
# with the help of this parameter we can make an column as an index
df5=pd.read_csv("D:\\8th_semester\\research data\\archive\\train.csv", index_col='POSTED_BY' ,usecols=[0,1,2,3], skiprows=[15,14,16,17,18])
print(df5)
# inthis case POSTED_BY column is index of this DataFrame
```

```python
# using index number of column
# with the help of this parameter we can make an column as an index index_col= index number
# RERA is at index 2
df5=pd.read_csv("D:\\8th_semester\\research data\\archive\\train.csv", index_col=2 ,usecols=[0,1,2,3], skiprows=[15,14,16,17,18])
print(df5)
# in this case RERA column is index of this DataFrame
```
