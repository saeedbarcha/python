# .sin() using this function to find value

```python
a_sin=np.sin(180) # sine() function takes an argumrnt as angle value to sind its sine
print(a_sin)
```

# creating ma matrix contains values

# creating x values

```python
x_sin=np.arange(0,3*np.pi, 0.1)
print(x_sin)
```

# creating y values

```python
y_sin=np.sin(x_sin)
print(y_sin)
```

# we will use x and y values to draw a graph matplotlib 

# sin() curv

```python
# .plot(x,y) takes tow argumrnts x and y
```

```python
plt.plot(x_sin, y_sin) 
plt.show()
```

# cos() curv

```python
y_cos=np.cos(x_sin) #to convert x_sin tables values into cos
plt.plot(x_sin, y_cos)
plt.show()
```

# tan() curv

```python
y_tan=np.tan(x_sin) #to convert x_sin tables values into cos
plt.plot(x_sin, y_tan)
plt.show()
```
