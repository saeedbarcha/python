```python
# importing modules
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
import random
```

```python
# creating an image using plt.pie() function
style.use("ggplot")
plt.figure(figsize=(16,9))
label=["Python","Java","JavaScript","HTML","CSS","Android"] #labels
color=["m","r","g","b","k","y"]
myfig=plt.pie([1,2,3,4,5,6], labels=label, colors=color, autopct="%.1f%%")
plt.legend()
plt.show(myfig)
```

# dpi=100 parameter to control resolution

```python
# with dpi=100
plt.figure(figsize=(16,9))
plt.pie([1,2,3,4,5,6], labels=label, colors=color, autopct="%.1f%%")
plt.legend()
#      dpi stands for density per/inch     
plt.savefig("chart1", dpi=100)       
plt.show()
```

# dpi=1000 parameter to control resolution

```python
# with dpi=1000
plt.figure(figsize=(16,9))
plt.pie([1,2,3,4,5,6], labels=label, colors=color, autopct="%.1f%%")
plt.legend()   
plt.savefig("chart2", dpi=1000)       
plt.show()
```
