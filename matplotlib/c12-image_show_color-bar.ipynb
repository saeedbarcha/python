# to increase the size of the image

```python
plt.figure(figsize=(16,9)) # to increase the size of the image
plt.axis("off") 
plt.imshow(img)
plt.show()
```

# to show the color bar of an image

```python
plt.figure(figsize=(16,9)) 
plt.axis("off") 
plt.imshow(img)
plt.colorbar() #this function is use to show color bar
               #we have to put this function next to imshow() function  
plt.show()
```
