```python
# to use time module we have to import time module first .this is built-in module
import time

# for example 
# i want to find the time taken for execution by for and while loop
# we can do this using time module
# for i in range(100):
#     print("hello world")
   

# print("#####################################################################################################################")

# k=0
# while(k<=100):
#     print("hello world")
#     k+=1



# solution
# creating a variable and assign present time
initial=time.time()

for i in range(10000):
    print(f"hello world: {i}")
a=f"time taken by for loop is {time.time()-initial} seconds"
print(f"time taken by for loop is {time.time()-initial} seconds")
initial2=time.time()

k=0
while(k<=10000):
    print("hello world")
    k+=1
    
print(f"time taken by while loop is {time.time()-initial2} seconds")
b=f"time taken by while loop is {time.time()-initial2} seconds" 


print(a)
print(b)
```

```python
# to print a present date in human understandable format
# we have method in time module
# time.time() returns  number of tics
# time.localtime() this will convert number of tics in to local time format
# time.asctime() convert it into presentable time format
localtime=time.asctime(time.localtime(time.time()))
print(localtime)
```

```python
# time.sleep() this is use to wait for a limited period of time
listi=["s","f","gg","kk"]
for i in listi:
    print(i)
    time.sleep(5)
```
